```python
import pandas as pd
import numpy as np 
```

```python
##look at expenses 
expenses = pd.read_csv(r'LAPoliceExpensesz.csv')
expenses.head()
```

```python
expenses['MONTH_YEAR'] = pd.to_datetime(expenses['FISCAL YEAR'].astype(str) + '-' + expenses['CALENDAR MONTH NUMBER'].astype(str))
```

```python
expenses['month_year'] = expenses['MONTH_YEAR'].dt.to_period('M')
```

```python
##multiple 
```

```python
##test = 
expenses['ACCOUNT NAME'].unique()
```

1. “EQUIP” – Any and all Equipment expenses
2. “TRAIN” - Any and all Training expenses
3. “VEHI” – Any and all Vehicle expenses including maintenance
4. “OFFICE” – Any and all office and administrative expenses (often called “furniture and fixtures”)
5. “POLICE” – Any expenses using the classification Police
6. “TECH” – All technology expenses
7. “TRANSPO” – All transportation expenses

```python
equip = expenses['ACCOUNT NAME'].str.contains('EQUIP')
train = expenses['ACCOUNT NAME'].str.contains('TRAIN')
vehi = expenses['ACCOUNT NAME'].str.contains('VEH')
##off = expenses['ACCOUNT NAME'].str.contains('OFF')
```

```python
##OFFICE AND FURN
expenses[(expenses['ACCOUNT NAME'].str.contains('FURN')) | (expenses['ACCOUNT NAME'].str.contains('OFF'))]['ACCOUNT NAME'].unique()
```

```python
pol = expenses['ACCOUNT NAME'].str.contains('POL')
tech = expenses['ACCOUNT NAME'].str.contains('TECH')
trans = expenses['ACCOUNT NAME'].str.contains('TRANS')
```

1. First filter by each month and group 
2. Group by month year and aggregate the sum 
3. get new dataframe, and reset index
4. drop unwanted columns 
5. repeat for all 7 various filters 

6. append to last dataframe

```python
e = expenses[equip]['ACCOUNT NAME'].unique()
t = expenses[trans]['ACCOUNT NAME'].unique()
```

```python
#modify equip list without trans
new_equip = [x for x in e if x not in t]
```

```python
new_equip
```

```python
df_equip = expenses[expenses['ACCOUNT NAME'].isin(new_equip)].groupby('month_year').sum()
df_train = expenses[train].groupby('month_year').sum()
df_vehi = expenses[vehi].groupby('month_year').sum() #missing
df_off = expenses[(expenses['ACCOUNT NAME'].str.contains('FURN')) | (expenses['ACCOUNT NAME'].str.contains('OFF'))].groupby('month_year').sum()
df_pol = expenses[pol].groupby('month_year').sum() ##missing
df_tech = expenses[tech].groupby('month_year').sum()
df_trans = expenses[trans].groupby('month_year').sum() ##missing
```

```python
df_equip.reset_index(inplace=True)
df_train.reset_index(inplace=True)
df_vehi.reset_index(inplace=True)
df_off.reset_index(inplace=True)
df_pol.reset_index(inplace=True)
df_tech.reset_index(inplace=True)
df_trans.reset_index(inplace=True)
```

```python
df_equip.drop(['FISCAL YEAR','CALENDAR MONTH NUMBER'],axis=1,inplace=True)
df_train.drop(['FISCAL YEAR','CALENDAR MONTH NUMBER'],axis=1,inplace=True)
df_vehi.drop(['FISCAL YEAR','CALENDAR MONTH NUMBER'],axis=1,inplace=True)
df_off.drop(['FISCAL YEAR','CALENDAR MONTH NUMBER'],axis=1,inplace=True)
df_pol.drop(['FISCAL YEAR','CALENDAR MONTH NUMBER'],axis=1,inplace=True)
df_tech.drop(['FISCAL YEAR','CALENDAR MONTH NUMBER'],axis=1,inplace=True)
df_trans.drop(['FISCAL YEAR','CALENDAR MONTH NUMBER'],axis=1,inplace=True)
```

```python
df_equip.rename(columns={"VALUE OF SPEND":"EQUIPMENT_EXPENSES"},inplace=True)

df_train.rename(columns={"VALUE OF SPEND":"TRAINING_EXPENSES"},inplace=True)

df_vehi.rename(columns={"VALUE OF SPEND":"VEHICLE_EXPENSES"},inplace=True)

df_off.rename(columns={"VALUE OF SPEND":"OFFICE_EXPENSES"},inplace=True)

df_pol.rename(columns={"VALUE OF SPEND":"POLICE_CLASSIFIED_EXPENSES"},inplace=True)

df_tech.rename(columns={"VALUE OF SPEND":"TECH_EXPENSES"},inplace=True)

df_trans.rename(columns={"VALUE OF SPEND":"TRANSPORTATION_EXPENSES"},inplace=True)
```

```python
df_vehi.head()
```

```python
##join all frames 
##df_vehi * df_pol

##first includes train and pol
first = pd.merge(df_equip,df_pol,how='left', on='month_year')
```

```python
##second, add the vehic data
second = pd.merge(first,df_vehi, how='left',on='month_year')
```

```python
third = pd.merge(second,df_trans, how='left',on='month_year')
```

```python
fourth = pd.merge(third,df_train, how='left',on='month_year')
fifth = pd.merge(fourth,df_off, how='left',on='month_year')
```

```python
final1 = pd.merge(fifth,df_tech,how='left',on='month_year')
```
